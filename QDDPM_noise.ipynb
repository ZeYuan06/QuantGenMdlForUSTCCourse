# functions

```python
def corrNoiseEnsemble(psi0, delta, p, Ndata, seed):
    '''
    generate random states by applying exp(-i\epsilon XX) with probability p
    or exp(-i\epsilon ZZ) with probability 1-p on \psi_0
    \epsilon \in [-\delta, \delta]
    '''
    np.random.seed(seed)
    noise_types = np.random.choice([1, 2], size=Ndata, replace=True, p=[p, 1-p])
    X = np.array([[0, 1.], [1., 0]])
    Z = np.array([[1., 0], [0, -1.]])
    XX = contract('ij,kl->ikjl', X, X).reshape((4, 4))
    ZZ = contract('ij,kl->ikjl', Z, Z).reshape((4, 4))
    angles = np.random.uniform(-delta, delta, Ndata).reshape((Ndata, 1, 1))
    xx_angles = angles[noise_types==1]
    zz_angles = angles[noise_types==2]
    Rxx = np.cos(xx_angles)*np.eye(4) - 1j*np.sin(xx_angles)*XX
    Rzz = np.cos(zz_angles)*np.eye(4) - 1j*np.sin(zz_angles)*ZZ
    Us = np.vstack((Rxx, Rzz))
    states = contract('mij, j->mi', Us, psi0)
    return states
```

# Diffusion process

## generate samples through diffusion

```python
# generate noisy ensemble
n = 2
T = 20
Ndata = 5000
delta = np.pi/3.
p = 0.8

diff_hs = np.hstack((np.linspace(0.05, 1., 12), np.linspace(1.2, 4., 8))).astype(np.float32)

psi0 = np.array([np.sqrt(0.1)*np.exp(1j*0.8), np.sqrt(0.55)*np.exp(1j*1.7), 
                    0., np.sqrt(0.35)*np.exp(1j*2.4)])
X = tf.convert_to_tensor(corrNoiseEnsemble(psi0, delta, p, Ndata, seed=15))

Xout = np.zeros((T+1, Ndata, 2**n), dtype=np.complex64)
Xout[0] = X

model_diff = MultiQubitDiffusionModel(n, T, Ndata)
for t in range(1, T+1):
    Xout[t] = model_diff.set_diffusionDataMulti_t(t, X, diff_hs[:t], seed=t)
    print(t)
np.save('data/noise/corrNoiseDiff_n2T_N5000.npy', Xout)
```

# Backward training

```python
def Training_t(model, t, inputs_T, params_tot, Ndata, epochs):
    '''
    training for the backward PQC at step t using whole dataset
    Args:
    model: QDDPM model
    t: diffusion step
    params_tot: collection of PQC parameters for steps > t 
    Ndata: number of samples in training data set
    epochs: number of iterations
    '''
    input_tplus1 = model.prepareInput_t(inputs_T, params_tot, t, Ndata) # prepare input
    states_diff = model.states_diff
    loss_hist = [] # record of training history

    # initialize parameters
    tf.random.set_seed(None)
    params_t = tf.Variable(tf.random.normal([2 * model.n_tot * model.L]))
    # set optimizer and learning rate decay
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0005)

    t0 = time.time()
    for step in range(epochs):
        indices = np.random.choice(states_diff.shape[1], size=Ndata, replace=False)
        true_data = states_diff[t, indices]
        with tf.GradientTape() as tape:
            output_t = model.backwardOutput_t(input_tplus1, params_t)
            loss = naturalDistance(output_t, true_data)
        grads = tape.gradient(loss, [params_t])
        optimizer.apply_gradients(zip(grads, [params_t]))

        loss_hist.append(tf.stop_gradient(loss)) # record the current loss
        
        if step%100 == 0:
            loss_value = loss_hist[-1]
            print("Step %s, loss: %s, time elapsed: %s seconds"%(step, loss_value, time.time() - t0))
        
    return tf.stop_gradient(params_t), tf.squeeze(tf.stack(loss_hist))
```

```python
n, na = 2, 2
T = 20
L = 6
Ndata = 500
epochs = 1001

# Prefer reusing existing trained results if available
params_all_path = f'data/noise/QDDPMcorrNoiseparams_n{n}na{na}T{T}L{L}_mmd.npy'
loss_all_path = f'data/noise/QDDPMcorrNoiseloss_n{n}na{na}T{T}L{L}_mmd.npy'
params_step_path = lambda t: f'data/noise/QDDPMcorrNoiseparams_n{n}na{na}T{T}L{L}_t{t}_mmd.npy'
# historical naming in this notebook is inconsistent; support both
loss_step_path_candidates = lambda t: [
    f'data/noise/QDDPMcorrNoiseloss_n{n}na{na}T{T}L{L}_t{t}_mmd.npy',
    f'data/noise/QDDPMcorrNoiselosshist_n{n}na{na}T{T}L{L}_t{t}_mmd.npy',
]

expected_param_dim = 2 * (n + na) * L

def _load_first_existing(paths):
    for p in paths:
        if os.path.exists(p):
            return np.load(p), p
    return None, None

# If combined results already exist and look sane, skip training
if os.path.exists(params_all_path) and os.path.exists(loss_all_path):
    _params_all = np.load(params_all_path)
    _loss_all = np.load(loss_all_path)
    ok_shape = (_params_all.shape == (T, expected_param_dim)) and (_loss_all.shape[0] == T)
    ok_finite = np.isfinite(_params_all).all() and np.isfinite(_loss_all).all()
    if ok_shape and ok_finite:
        print(f"[skip] Found trained results: {params_all_path} and {loss_all_path}")
        print(f"        params shape={_params_all.shape}, loss shape={_loss_all.shape}")
    else:
        print("[warn] Found combined files but they look inconsistent; will (re)train missing steps.")
        print(f"       params shape={_params_all.shape}, loss shape={_loss_all.shape}, ok_shape={ok_shape}, ok_finite={ok_finite}")
        # fall through to training below
else:
    # fall through to training below
    pass

# Train only if we don't have a good combined checkpoint
need_train = True
if os.path.exists(params_all_path) and os.path.exists(loss_all_path):
    _params_all = np.load(params_all_path)
    _loss_all = np.load(loss_all_path)
    need_train = not ((_params_all.shape == (T, expected_param_dim)) and (_loss_all.shape[0] == T) and np.isfinite(_params_all).all() and np.isfinite(_loss_all).all())

if need_train:
    model_diff = MultiQubitDiffusionModel(n, T, Ndata)
    inputs_T = model_diff.HaarSampleGeneration(Ndata, seed=22)

    model = QDDPM(n=n, na=na, T=T, L=L)
    states_diff = np.load('data/noise/corrNoiseDiff_n%dT%d_N5000.npy'%(n, T))
    model.set_diffusionSet(states_diff)

    for t in range(T-1, -1, -1):
        # If this step already has parameters and a loss history saved, reuse them
        has_params_t = os.path.exists(params_step_path(t))
        loss_each, loss_path = _load_first_existing(loss_step_path_candidates(t))
        if has_params_t and (loss_each is not None):
            print(f"[skip] step t={t}: found {params_step_path(t)} and {loss_path}")
            continue

        params_tot = np.zeros((T, expected_param_dim), dtype=np.float32)
        # load already-available params for steps > t
        for tt in range(t+1, T):
            if os.path.exists(params_step_path(tt)):
                params_tot[tt] = np.load(params_step_path(tt))
            else:
                # if missing, keep zeros; Training_t will still run but may degrade quality
                pass

        params, loss = Training_t(model, t, inputs_T, params_tot, Ndata, epochs)
        np.save(params_step_path(t), params.numpy() if hasattr(params, 'numpy') else params)
        np.save(f'data/noise/QDDPMcorrNoiseloss_n{n}na{na}T{T}L{L}_t{t}_mmd.npy', loss.numpy() if hasattr(loss, 'numpy') else loss)
        print(f"[saved] step t={t}: params and loss")
```

## generate training and testing data

```python
# collect optimized results (skip if already collected)
n, na = 2, 2
T = 20
L = 6
Ndata = 500

params_all_path = f'data/noise/QDDPMcorrNoiseparams_n{n}na{na}T{T}L{L}_mmd.npy'
loss_all_path = f'data/noise/QDDPMcorrNoiseloss_n{n}na{na}T{T}L{L}_mmd.npy'
params_step_path = lambda t: f'data/noise/QDDPMcorrNoiseparams_n{n}na{na}T{T}L{L}_t{t}_mmd.npy'
loss_step_path_candidates = lambda t: [
    f'data/noise/QDDPMcorrNoiseloss_n{n}na{na}T{T}L{L}_t{t}_mmd.npy',
    f'data/noise/QDDPMcorrNoiselosshist_n{n}na{na}T{T}L{L}_t{t}_mmd.npy',
]

expected_param_dim = 2 * (n + na) * L

def _load_first_existing(paths):
    for p in paths:
        if os.path.exists(p):
            return np.load(p), p
    raise FileNotFoundError(f"None of these files exist: {paths}")

if os.path.exists(params_all_path) and os.path.exists(loss_all_path):
    params_tot = np.load(params_all_path)
    loss_tot = np.load(loss_all_path)
    print(f"[skip] Using existing collected results: {params_all_path}, {loss_all_path}")
    print(f"       params shape={params_tot.shape}, loss shape={loss_tot.shape}")
else:
    params_tot = np.zeros((T, expected_param_dim))
    loss_list = []
    for t in range(T):
        params_tot[t] = np.load(params_step_path(t))
        loss_each, loss_path = _load_first_existing(loss_step_path_candidates(t))
        loss_list.append(loss_each)
        if t == 0:
            print(f"[collect] example loss file: {loss_path}, shape={loss_each.shape}")

    # loss histories can have varying lengths; store as object array to be safe
    loss_tot = np.array(loss_list, dtype=object)
    np.save(loss_all_path, loss_tot)
    np.save(params_all_path, params_tot)
    print(f"[saved] {params_all_path}")
    print(f"[saved] {loss_all_path}")
```
